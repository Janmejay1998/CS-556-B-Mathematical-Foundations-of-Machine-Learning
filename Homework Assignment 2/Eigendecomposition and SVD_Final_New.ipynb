```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define some matrix `a`
a = [[1, 5], [2,4]]
```

```python
evals,evecs = np.linalg.eig(a)
```

```python
print("Eigen Values:\n {}".format(evals))
print("\nEigen Vectors:\n {}".format(evecs))
```

```python
np.random.seed(123)
v1 = np.array([1,1])
# v1 = v1/np.linalg.norm(v1)
v2 = np.random.rand(2,1)
v2 = v2/np.linalg.norm(v2)#normalize it to unit vector
```

```python
av1 = a@v1
av2 = a@v2
```

```python
fig,axs=plt.subplots(1,2,figsize=(20,9))
ax=axs[0]
ax.plot([0, v1[0]], [0, v1[1]], 'k',lw=6)
ax.plot([0, av1[0]], [0, av1[1]], 'r--',lw=2)
ax.set_title("Vector $\mathbf{v}_1$ Before & After\nCoordinate Transform ()",fontsize=24)
ax.tick_params(axis='both', which='major', labelsize=20)

ax=axs[1]
ax.plot([0, v2[0]], [0, v2[1]], 'k',lw=6)
ax.plot([0, av2[0]], [0, av2[1]], 'r--',lw=2)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set_title("Vector $\mathbf{v}_2$ Before & After\nCoordinate Transform",fontsize=24)
```

# Singular Value Decompositon

```python
a = np.array([[4, 0], [3, -5]]) #Some random matrix of real values `a` , note: `a` can be square or rectangular
u,s,v = np.linalg.svd(a)
print("U =\n {}\n\nS =\n {}\n\nV =\n {}".format(u,np.diag(s),v))
```

```python
print("Original Matrix A\n {}".format(a))
print("\nSVD Approximation\n {}".format(np.matmul(u,np.matmul(np.diag(s), v))))
```

```python
#v is `Orthonormal`
print("vTv = I => \n{}".format(np.matmul(v.T,v)))

#u is `Orthonormal`
print("\nuTu = I => \n{}".format(np.matmul(u.T,u)))
```
